```python
from langchain_core.language_models import BaseChatModel
from langchain_core.messages import AIMessage, HumanMessage, SystemMessage
from langchain_core.outputs import ChatResult, ChatGeneration
from typing import List, Dict, Any
import requests


class DeepSeekChat(BaseChatModel):
    api_url: str = "https://api.intelligence.io.solutions/api/v1/chat/completions"
    api_key: str = "Bearer io-v2-eyJhbGciOiJSUzI1NiIsInR5cCI6IkpXVCJ9.eyJvd25lciI6IjkwNTlkMjY4LTJiNzItNDkwOS1hMzhkLWNiYmFjZGNjOTE4NiIsImV4cCI6NDg5NTcyNTc4NX0.Goaob-cEAtr_DaQu19GGM-e33Znj95aGlvVJz0GiRij2MWVfgiFHWjRVWRMf4CosBBqAxvpBMfU0WRplNMxXtg"
    model: str = "deepseek-ai/DeepSeek-R1"
    temperature: float = 0.3

    def _convert_messages(self, messages: List) -> List[Dict]:
        """Конвертирует сообщения LangChain в формат API"""
        converted = []
        for msg in messages:
            if isinstance(msg, HumanMessage):
                role = "user"
            elif isinstance(msg, AIMessage):
                role = "assistant"
            elif isinstance(msg, SystemMessage):
                role = "system"
            else:
                raise ValueError(f"Unsupported message type: {type(msg)}")

            converted.append({"role": role, "content": msg.content})
        return converted

    def _generate(self, messages: List, **kwargs) -> ChatResult:
        time.sleep(1.5)
        """Генерация ответа с явным системным промптом"""
        # 1. Добавляем системный промпт к сообщениям
        system_prompt = SystemMessage(
            content="""Ты AI для анализа данных. Всегда отвечай СТРОГО в формате:
Thought: [анализ задачи]
Action: python_repl_ast

Пример:
Thought: Проверю пустые значения
Action: python_repl_ast
Action Input: df.isnull().sum()"""
        )
        all_messages = [system_prompt] + messages

        # 2. Отправка запроса
        headers = {"Content-Type": "application/json", "Authorization": self.api_key}
        response = requests.post(
            self.api_url,
            headers=headers,
            json={
                "model": self.model,
                "messages": self._convert_messages(all_messages),
                "temperature": self.temperature,
                **kwargs
            }
        )
        response.raise_for_status()

        # 3. Обработка ответа
        result = response.json()
        content = result['choices'][0]['message']['content']

        #4. Принудительное форматирование при необходимости
        # if not content.startswith("Thought:"):
        #     content = f"Thought: {content}\nAction: python_repl_ast\nAction Input: df.isnull().sum()"

        return ChatResult(
            generations=[ChatGeneration(message=AIMessage(content=content))]
        )

    @property
    def _llm_type(self) -> str:
        return "deepseek-r1"
```

```python
agent = create_pandas_dataframe_agent(
    llm=DeepSeekChat(),
    df=df_for_LLM,
    verbose=True,
    allow_dangerous_code=True,
    handle_parsing_errors=lambda e: f"Ошибка парсинга: {str(e)[:50]}",
    max_iterations=20
)
```
